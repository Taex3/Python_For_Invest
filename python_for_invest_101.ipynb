```python
import pandas as pd
import mplfinance as mpf
from binance.client import Client
import yfinance as yf
```

Prepare Data

```python
start = '2020-01-01'
end = '2021-01-01'
```

```python
data = yf.download("GC=f", period='1d', start=start, end=end)
```

```python
data
```

```python
client = Client()
```

```python
data = client.get_historical_klines("ETHUSDT", Client.KLINE_INTERVAL_1DAY, start, end)
```

```python
data
```

```python
ETH = pd.DataFrame(data, columns =['otime','open','high','low','close','volume','ctime',
                                'quote_vol','no_trade','taker_base_vol','taker_quote_vol','ignore'])
```

```python
ETH
```

```python
ETH = ETH[['otime','open','high','low','close','volume']] #[[]] ---> select columns
ETH['otime'] = pd.to_datetime(ETH['otime'], unit = 'ms')
```

```python
ETH['otime'] #[] -----> output = series and can use only 1columns
```

```python
ETH[['otime']] # ------> output = dataframe
```

```python
ETH
```

```python
ETH = ETH.set_index('otime')
```

```python
ETH = ETH.astype(float)
```

Make add plot

```python
ret
```

```python
ret_add = mpf.make_addplot(ret, color = 'green', ylabel = 'return')
```

```python
mpf.plot(ETH, addplot = ret_add)
```

```python
ret_add = mpf.make_addplot(ret, color = 'green', panel = 1, ylabel = 'return')
```

```python
mpf.plot(ETH, addplot = ret_add) # -----> add panel = 1
```

```python
mpf.plot(ETH, addplot = ret_add, volume = True)
```
